```python
#Code block
# Game board
board = [
    ['#', '_', '_', '_', '_', '#', '_'],
    ['_', '#', '#', '_', '#', '#', '_'],
    ['_', '#', '#', '_', '#', '#', '_'],
    ['_', '#', '_', '_', '_', '_', '_'],
    ['#', '#', '_', '#', '#', '#', '#'],
    ['_', '_', '_', '_', '_', '_', '_'],
    ['#', '#', '_', '#', '#', '#', '#'],
    ['_', '_', '_', '_', '_', '_', '#'],
    ['#', '#', '_', '#', '#', '#', '#'],
    ['#', '#', '_', '_', '_', '_', '#'],
]
words = ['DUCK', 'DOVE', 'PEACOCK', 'PARROT', 'SWANE','WREN','CROW', 'EMU', 'SPARROW']
myMap = dict({'DUCK' : 0 ,'DOVE' : 0 , 'PEACOCK' : 0 ,'PARROT' : 0 ,'SWANE' : 0 ,'WREN' : 0 ,'CROW' : 0 ,'EMU' : 0 ,'SPARROW' : 0})
scores_g = None
```

Choice and implementation of the Static Evaluation Function.

```python
#Code block
# Function to evaluate score
def evaluate(board):
    score = 0
    for row in board:
        for char in row:
            if char != '_':
                score += 1
    #print("evaluate" , score)
    return score
```

### Implementation of the Min-Max algorithm

```python
#Code block
# Function to check if a word can be placed at a particular position
MAX = 99999999
MIN = -99999999
def can_place_word(word, x, y, direction):
    if direction == 'across':
        # Check if word fits within the board horizontally
        if y + len(word) > len(board[0]):
            return False
        # Check if word intersects with any existing letters
        for i, char in enumerate(word):
            #print("before err x , y , i , len_board0 , len_board , len_word" , x , y , i , len(board[0]) ,len(board) , word)
            #print("chars" ,board[x][y+i] )
            if board[x][y+i] != '_' and board[x][y+i] != char:
                return False
    elif direction == 'down':
        # Check if word fits within the board vertically
        if x + len(word) > len(board):
            return False
        # Check if word intersects with any existing letters
        #print("in can_place_word" , word , x , y)
        for i, char in enumerate(word):
            #print("before err" , x , y , i , len(board))
            if board[x+i][y] != '_' and board[x+i][y] != char:
                return False
    return True

# Function to update the game board with a word
def update_board(word, x, y, direction , undo):
    #print(" update_board word", word)
    t_word = ''
    #if undo:
        #print(word)
    if direction == 'across':
        for i, char in enumerate(word):
            #if undo:
            #    char = '_'
                #print("across before char" , board[x][y+1] , char )
            t_word = t_word + board[x][y+i]
            board[x][y+i] = char
            
    elif direction == 'down':
        for i, char in enumerate(word):
            #if undo:
            #    char = '_'
                #print("down before char" , board[x+i][y] , char)
            t_word = t_word + board[x+i][y]
            board[x+i][y] = char
    #print("t_word" , t_word)
    return t_word

# Function to get all possible moves for a player
def get_possible_moves(board, words):
    moves = []
    for word in words:
        #print(" get_possible_moves word" , word , myMap[word])
        if myMap[word] == 1 :
            continue
        for i in range(len(board)):
            for j in range(len(board[0])):
                if can_place_word(word, i, j, 'across'):
                    #print("possible move 1 " , word , i , j )
                    moves.append((word, i, j, 'across'))
                if can_place_word(word, i, j, 'down'):
                    #print("possible move 2 " , word , i , j )
                    moves.append((word, i, j, 'down'))
    #print("get_possible_moves return" ,moves )
    #print("len of possible moves" , len(moves))
    return moves

# Min-Max algorithm to decide the best move for a player
def minimax(board, words,alpha , beta , depth,maximizing_player):
    if depth == 0:
        return None, evaluate(board)
    
    global myMap
    global scores_g
    if maximizing_player:
        #print("maximizing_player")
        best_move = None
        word_out = None
        x_out = None
        y_out = None 
        direction_out = None
        best_score = float('-inf')
        #print("maximizing_player before get_possible_moves")
        for move in get_possible_moves(board, words):
            # Make a move and evaluate the score
            #print("maximize player move" , move)
            word, x, y, direction = move
            word_out = word
            x_out = x
            y_out = y
            prev_word = ''
            direction_out = direction 
            #print("maximizing before update_board" , word)
            myMap[word] = 1 
            prev_word = update_board(word, x, y, direction, 0) 
            #print("prev_word" , prev_word)
            score = minimax(board, words, alpha , beta , depth-1, False)[1]
            # Undo the move
            myMap[word] = 0
            #print("maximizing before undo update_board" , word)
            update_board(prev_word, x, y, direction , 1)
            #Update the best move and score
            if score > best_score:
                best_move = move
                best_score = score
                scores_g = score
            alpha = max(alpha  , best_score)
            #if beta <= alpha:
            #    break
        #print("here 2" , word, x, y, direction)
        return best_move, best_score, word_out, x_out, y_out, direction_out 
    else:
        #print("minimizing_player before get_possible_moves")
        best_move = None
        word_out = None
        x_out = None
        y_out = None 
        prev_word = ''
        direction_out = None
        best_score = float('inf')
        for move in get_possible_moves(board, words):
            # Make a move and evaluate the score
            #print("minimize player move" , move)
            word, x, y, direction = move
            word_out = word
            x_out = x
            y_out = y
            direction_out = direction 
            #print("minimizing before  update_board" , word)
            myMap[word] = 1
            prev_word = update_board(word, x, y, direction , 0)
            #print("prev_word" , prev_word)
            score = minimax(board, words, alpha, beta , depth-1,True)[1]
            # Undo the move
            myMap[word] = 0
            #print("minimizng before undo update_board" , word)
            update_board(prev_word, x, y, direction , 1)
            # Update the best move and score
            if score <= best_score:
                best_move = move
                best_score = score
                scores_g = score
            beta = min(beta, best_score)
            #if beta <= alpha:
            #    break
        #print("here 1", word, x, y, direction)
        return best_move, best_score , word_out, x_out, y_out, direction_out 
```
